# Escala de Tons de Cinza

```python
from sklearn.datasets import load_sample_image
import matplotlib.pyplot as plt
import numpy as np
```

```python
china = load_sample_image("china.jpg")
```

```python
plt.imshow(china)
plt.axis('off')  # remove os eixos
plt.show()
```

```python
china_red = china[:,:,0]
china_green = china[:,:,1]
china_blue = china[:,:,2]
```

```python
gray_china = 0.299 * china_red + 0.587 * china_green + 0.114 * china_blue
```

```python
plt.imshow(np.stack((gray_china, gray_china, gray_china), axis=2).astype(np.uint8))
plt.axis('off')  # remove os eixos
plt.show()
```
